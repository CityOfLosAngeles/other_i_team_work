```python
full_streets = pd.read_csv('clean_stat_segment_new_project.csv')
score_columns = full_streets.columns[['Scor' in x for x in full_streets.columns]].values
clean_summaries = pd.concat([full_streets.pivot_table('length_2',index='CDNum',columns=x,aggfunc='sum')for x in score_columns],keys=score_columns)
```

```python
def summarizer(quarter_list,total=True,cds=None):
    items = []
    for item in quarter_list:
        if total == True:
            items.append(clean_summaries.loc[item].sum().values)
            
        else:
            items.append(clean_summaries.loc[item].loc[cds].sum())
        
    df = pd.DataFrame(items,columns=[0,1,2,3],index=score_columns)
    return df.rename(columns={0:'Not Eval',1:'Clean',2:'Somewhat Clean',3:'Not Clean'})[['Not Clean','Somewhat Clean','Clean']]
```
